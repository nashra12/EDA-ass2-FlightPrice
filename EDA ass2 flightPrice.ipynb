Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset
to identify the peak travel season. What features would you analyze to identify the peak season, and how

```python
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
```

```python
df.head(2)
```

```python
df['Month'].unique()
```

```python
import matplotlib.pyplot as plt
# Plot the average price per month
monthly_avg_prices = df.groupby('Month')['Price'].mean()
monthly_avg_prices.plot(kind='line', marker='o')
plt.title('Average Flight Price per Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
```

Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight
Price dataset to identify any trends in flight prices. What features would you analyze to identify these
trends, and what visualizations would you use to present your findings to your team?

```python
df.head(2)
```

```python
# Convert Date_of_Journey to datetime
# df['Date_of_Journey'] = pd.to_datetime(flights_df['Date_of_Journey'], format='%d/%m/%Y')
```

```python
# 2. Box Plot for Prices by Airline
plt.figure(figsize=(12, 8))
sns.boxplot(x='Price', y='Airline', data=df)
plt.title('Flight Prices by Airline')
plt.xlabel('Price')
plt.ylabel('Airline')
plt.show()
```

```python
# 3. Bar Chart for Average Prices by Number of Stops
stops_avg_prices = df.groupby('Total_Stops')['Price'].mean()
plt.figure(figsize=(10, 6))
stops_avg_prices.plot(kind='bar')
plt.title('Average Flight Prices by Number of Stops')
plt.xlabel('Number of Stops')
plt.ylabel('Average Price')
plt.show()
```

```python
Destination_avg_prices = df.groupby('Destination')['Price'].mean()
plt.figure(figsize=(10, 6))
Destination_avg_prices.plot(kind='bar')
plt.title('Average Flight Prices by Destination')
plt.xlabel('Destination')
plt.ylabel('Average Price')
plt.show()
```

Q8. You are a data scientist working for an airline company, and you have been asked to analyze the
Flight Price dataset to identify the factors that affect flight prices. What features would you analyze to
identify these factors, and how would you present your findings to the management team?

```python
df.head(2)
```

```python
df['Hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Minute']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head(2)
```

```python
df=df.drop(['Date_of_Journey','Duration'],axis=1)
```

```python
df.head(2)
```

```python
df_corr_matrix=df[['Month','Date','Hour','Minute']].corr()
```

```python
sns.heatmap(df_corr_matrix,annot=True,robust=True)
```
